```python
# Добавим в функцию удаление всех двойных кавычек
# И запись пустого списка в случае неудачи
def text_to_json(text):
  try:
    return json.loads(text.replace('"', '').replace("'", '"'))
  except:
    return []
df3['key_skills_json'] = df3['key_skills'].apply(text_to_json)
```

```python
df3['key_skills_json']
```

```python
# возвращаем пустой список в случае каких-либо проблем
def get_values(json):
  try:
    out = []
    if len(json)>0:
      for i in json:
        out.append(i['name'])
    return out
  except:
    return []
df3['key_skills_list'] = df3['key_skills_json'].apply(get_values)
```

## Дедаем общий список из списков навыков

```python
sp1 = []
for i in range(0, len(df3['key_skills_list'])):
  sp1.append(df3['key_skills_list'][i])
sp1
new_list = []
def my_fun(temp_list):
    for ele in temp_list:
        if type(ele) == list:
            my_fun(ele)
        else:
            new_list.append(ele)
my_fun(sp1)
new_list
```

## Подсчитываем количество каждого навыка в списке и добавляем в словарь

```python
from collections import Counter
c = Counter(new_list)
c = dict(c)


# упорядочим наш словарь  порядке убывания
sorted_result = {k: c[k] for k in sorted(c, key=c.get, reverse=True)}
print(sorted_result)
```

## Строим гистограмму топ 10 самых встречаемых навыков

```python
top_10_skills = dict(list(sorted_result.items())[:10]) # отбираем топ 10 
top_10_skills
```

```python

import matplotlib.pyplot as plt
plt.barh(list(top_10_skills.keys()), top_10_skills.values()) 
plt.title('Топ 10 навыков, упоминаемых в вакансиях')
plt.show()
```

### Определяем и строим гистограмму топ 20 работодаталей

```python
def company_employers(employer):
  res = eval(employer)
  return res['name']
```

```python
df3['company_employer'] = df3['employer'].apply(company_employers) # применяем функцию, которая возвращает имя компании из словаря по ключу 'name'
company_counts = df3['company_employer'].value_counts() # серия с количеством повторяющихся компаний (в порядке убывания)
companies_frame = company_counts.to_frame().reset_index() # переводим серию в датафрейм
companies_frame.rename(columns={'index': 'Имя компании', 'company_employer': 'Количество вакансий'}, inplace=True) # переименуем колонки, чтобы они были более понятны (дальше название колонок будет отображаться на графике)
top_10_companies_frame = companies_frame.head(20) # возьмем первые 20
top_10_companies_frame
```

```python
from typing import ValuesView
import plotly.express as px 
fig = px.bar(top_10_companies_frame, x='Имя компании', y='Количество вакансий', text_auto='.2s', width=1000, labels={'x': 'Имя компании', 'y':'Количество вакансий'}, title="Столбцовая диаграмма распределеиня компаний нанимателей")
fig.show()
```

## Формируем датафрейм и строим  гистограмму топ 20 должностей по зарплате

```python
vacancies_salaries = df3[['name', 'from']]
vacancies_salaries = vacancies_salaries[vacancies_salaries['from'] != 0].dropna()
vacancies_salaries = vacancies_salaries.groupby('name', as_index=False).median() # сгруппируем по должностям и выведем медианную зп для каждой должности
vacancies_salaries = vacancies_salaries.sort_values('from', ascending=False).head(20) # сортируем по убыванию зарплаты (от) и возьмем первые 20 должностей 
vacancies_salaries.rename(columns={'name': 'Должность', 'from': 'Зарплата'}, inplace=True) # добавим в название наших колонок смысла
```

```python
from typing import ValuesView
import plotly.express as px 
fig = px.bar(vacancies_salaries, y='Должность', x='Зарплата', text_auto='.2s', width=1200, height=800, title="Топ 20 должностей по зарплатам")
fig.show()
```
