# Предобработка и очистка данных после парсинга

## Изменение формата с object на dict  для столбца salary

```python
import ast

def str_to_dict(string):
  return ast.literal_eval(string)

df['fill_salary'] = df['salary'].fillna("{'from': 0,'to': 0, 'currency': 'no data', 'gross': 'no data'}")
df['clear_salary'] = df['fill_salary'].apply(str_to_dict)
df_fe = df[['name', 'clear_salary']]
df

# df_fe['clear_salary'][: 5]
type(df_fe['clear_salary'][0])
df
```

## Вывод столбцов из словаря ячеек clean_salary

```python
a = df.clear_salary.apply(pd.Series) #Очищенный и распакованный из словаря датафрейм
a
```

## Объединение двух датафреймов (который был + 4 очищенных столбаца с зп от, до, валюта, gross или net)



```python
df3 = df.join(a)
df3 #объединённый датафрейм исходный + распакованные из словая столбцы о зп
```

## Перевод иностранной валюты в рубли и вычет налога

```python
df3['from']=np.where((df3['gross']== True), df3['from']*0.87, df3['from'])
df3['to']=np.where((df3['gross']== True), df3['to']*0.87, df3['to'])
df3['from']=np.where((df3['currency']== "USD"), df3['from']*53.0, df3['from'])
df3['to']=np.where((df3['currency']== "USD"), df3['to']*53.0, df3['to'])
df3['from']=np.where((df3['currency']== "EUR"), df3['from']*57.0, df3['from'])
df3['to']=np.where((df3['currency']== "EUR"), df3['to']*57.0, df3['to'])
df3['from']=np.where((df3['currency']== "KZT"), df3['from']*0.12, df3['from'])
df3['to']=np.where((df3['currency']== "KZT"), df3['to']*0.12, df3['to'])
df3
```
